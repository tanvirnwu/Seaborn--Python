# Multi-plot Grid (FacetGrid Plot)
Multi-plot grids, also known as grid plots, are a type of visualization tool used to display multiple plots or graphs within a single figure or layout. They are often used in data analysis and scientific research to compare and contrast different data sets or to visualize multiple aspects of a single data set.

In a multi-plot grid, each individual plot is displayed as a separate panel within a larger grid or matrix. The panels can be arranged in rows and columns to create a structured layout that allows for easy comparison between different plots.

Multi-plot grids can be created using various software tools and programming languages such as Python's Matplotlib and Seaborn libraries, R's ggplot2, and MATLAB's subplot function. These tools provide a range of customization options, such as adjusting the size and position of each panel, adding labels and titles to the plots, and selecting different color schemes and styles.

Overall, multi-plot grids are a flexible and powerful visualization tool that allows researchers to explore and analyze complex data sets in a systematic and organized way.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('mart_linePlot.xlsx')
mart.columns= mart.columns.str.lower()
mart.head()
```

```python
mart.outlet_size.unique()
```

```python
mart.outlet_location_type.unique()
```

```python
sns.FacetGrid(data = mart, col='outlet_size')
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size')
myobj1.map_dataframe(sns.histplot, x='sales')
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size')
myobj1.map_dataframe(sns.lineplot, x='outlet_year', y ='sales')
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size')
myobj1.map_dataframe(sns.scatterplot, x='outlet_year', y ='sales', hue='outlet_location_type',marker='*', alpha = 0.5)
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size')
myobj1.map_dataframe(sns.scatterplot, x='outlet_year', y ='sales', hue='outlet_location_type',marker='*', s = 55)
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size', row='outlet_location_type')
myobj1.map_dataframe(sns.histplot,  x ='sales', hue='outlet_location_type')
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size', row='outlet_location_type', sharey = False)
myobj1.map_dataframe(sns.histplot,  x ='sales', hue='outlet_location_type')
plt.show()
```

```python
plt.figure(figsize=(15,8))

myobj1 = sns.FacetGrid(data = mart, col='outlet_size', row='outlet_location_type', sharey = False, ylim=(0,1000))
myobj1.map_dataframe(sns.histplot,  x ='sales', hue='outlet_location_type')
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size', row='outlet_location_type', sharey = False, ylim=(0,1000))
myobj1.map_dataframe(sns.histplot,  x ='sales', hue='outlet_location_type')
myobj1.set_axis_labels('sales','count')
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size', row='outlet_location_type', sharey = False, ylim=(0,1000))
myobj1.map_dataframe(sns.histplot,  x ='sales', hue='outlet_location_type')
myobj1.set_axis_labels('sales','count')
myobj1.set_titles(col_template = '{col_name}', row_template = '{row_name}')
plt.show()
```

```python
myobj1 = sns.FacetGrid(data = mart, col='outlet_size', row='outlet_location_type', sharey = False, ylim=(0,1000))
myobj1.map_dataframe(sns.histplot,  x ='sales', hue='outlet_location_type')
myobj1.set_axis_labels('sales','count')
myobj1.set_titles(col_template = '{col_name} Size', row_template = '{row_name} Type')
plt.show()
```
