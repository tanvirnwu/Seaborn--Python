# What is Relational Plot?
A relational plot is a type of data visualization that shows the relationship between two variables. It is used to examine the correlation or association between two variables in a dataset, and to explore the strength and direction of the relationship.

Relational plots are typically created using scatter plots, line plots, or other graphical representations. They can be used to explore the relationship between continuous variables, such as height and weight, or between categorical variables, such as gender and occupation.

## Scatter Plot
A scatter plot is a type of data visualization that shows the relationship between two variables. It is used to examine the correlation or association between two variables in a dataset, and to explore the strength and direction of the relationship.

In a scatter plot, each data point is represented as a point on a two-dimensional plane, where one variable is plotted along the x-axis and the other variable is plotted along the y-axis. The position of each point indicates the values of the two variables for that data point.

Scatter plots can help identify patterns in the data, such as a positive or negative correlation between the two variables. A positive correlation means that as one variable increases, the other variable also tends to increase. A negative correlation means that as one variable increases, the other variable tends to decrease.

Scatter plots are commonly used in data analysis and data visualization, and are often created using libraries such as matplotlib or Seaborn in Python, or ggplot in R. They are useful for identifying outliers, clusters of data points, and other patterns that may be present in the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel('mart_train_sample.xlsx')
mart.head()
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales')
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales', hue='Outlet_Size')
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales', style='Outlet_Size')
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales', style='Outlet_Size', markers={"High":"*", "Small":"v", "Medium":"^"})
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales', style='Outlet_Size',
                hue='Outlet_Size', s=80,
                markers={"High":"*", "Small":"v", "Medium":"^"})
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales', style='Outlet_Size',
                hue='Outlet_Location_Type', s=70,
                markers={"High":"*", "Small":"v", "Medium":"^"})
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales',
                #style='Outlet_Size',
                palette='Purples_r',
                hue='Outlet_Size', s=80,
                markers={"High":"*", "Small":"v", "Medium":"^"})
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y ='Sales',
                #style='Outlet_Size',
                palette='vlag',
                hue='Outlet_Size', s=100,
                size= 'Tier',
                sizes=(20,200),
                markers={"High":"*", "Small":"v", "Medium":"^"})
```
