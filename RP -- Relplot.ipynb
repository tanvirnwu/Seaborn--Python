# Relplot
A relplot is a type of data visualization in Python that is created using the seaborn library. It is used to explore the relationship between two numerical variables or one numerical and one categorical variable.

A relplot can be created using the seaborn function sns.relplot(). It allows you to create a scatter plot or a line plot, depending on the type of data you have. Additionally, it provides options to add additional dimensions to the plot, such as color, size, and style, to show the relationship between more than two variables.

A relplot is useful for exploring the correlation between two or more variables, identifying outliers or patterns in the data, and for visualizing trends over time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel('mart_train_sample.xlsx')
mart.head()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales')
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales', kind='line', ci=None)
plt.show()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales', s=75, hue='Outlet_Size')
plt.show()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales', s=75, hue='Outlet_Size',
            col = 'Item_Type')
plt.show()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales', s=75, hue='Outlet_Size',
            col = 'Item_Type', col_wrap=3)
plt.show()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales', s=75, hue='Outlet_Size',
            col = 'Outlet_Size', row='Tier')
plt.show()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y ='Sales', s=75, hue='Outlet_Size',
            col = 'Outlet_Size', row='Tier', palette='rocket_r')
plt.show()
```
