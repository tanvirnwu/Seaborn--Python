# Strip Plot
A strip plot is a type of plot in Seaborn that displays the distribution of data points as individual dots along a one-dimensional axis. It is a useful visualization tool for exploring the distribution of a dataset, especially when there are only a few data points.

In a strip plot, each data point is represented as a dot, and the dots are placed along the axis according to their value. The dots are often jittered, or slightly randomized, to avoid overlapping.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head(5)
```

```python
sns.stripplot(y='Total', data=mart)
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Payment', y='Total', data=mart)
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Payment', y='Total', hue='Gender' ,data=mart)
```

```python
sns.stripplot(y='Total', data=mart, jitter=0.2)
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(y='Total', data=mart, jitter=0.5)
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(y='Total', data=mart, jitter=0.2, linewidth=0.2, edgecolor='red')
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Payment', y = 'Total', data=mart, hue='Gender' , jitter=0.2, linewidth=0.5)
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Payment', y = 'Total', data=mart, hue='Gender' , jitter=0.2, linewidth=0.5, dodge=True)
plt.legend(loc='upper right')
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Payment', y = 'Total', data=mart, hue='Gender' , jitter=0.2, linewidth=0.5, dodge=True)
plt.legend(loc='upper right')
sns.violinplot(data=mart, y='Total', x='Payment', color='White')
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Payment', y = 'Total', data=mart, hue='Gender' , jitter=0.2, linewidth=0.5, dodge=True)
plt.legend(loc='upper right')
sns.boxplot(data=mart, y='Total', x='Payment', color='White')
```
