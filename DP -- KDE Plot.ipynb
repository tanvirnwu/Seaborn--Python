# KDE Plot
A KDE plot (Kernel Density Estimation plot) is a type of plot that is used to visualize the probability density function of a dataset. It is a non-parametric way to estimate the probability density function of a continuous random variable.

A KDE plot works by first placing a kernel, which is usually a Gaussian (normal) distribution, at each data point in the dataset. The kernel is then summed over all data points to create a smooth, continuous curve that estimates the probability density function of the data. The height of the curve at a particular point represents the estimated probability density at that point.

KDE plots are commonly used in data analysis and statistics to visualize the distribution of a dataset, and to identify any patterns or outliers. They can be used to compare the distributions of different datasets, or to examine the distribution of a single variable. KDE plots can also be used to estimate the probability density function of a random variable in cases where the true distribution is unknown or difficult to estimate.

In summary, a KDE plot is a useful tool for visualizing the distribution of a dataset and provides an estimate of the probability density function.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

```python
sns.kdeplot(data=mart, x='Total')
```

```python
sns.kdeplot(data=mart)
```

```python
sns.kdeplot(data=mart, x='Total',bw_adjust=0.5)
```

```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack', bw_adjust=0.5)
```

```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack', bw_adjust=0.5, log_scale=True)
```

```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack', bw_adjust=0.5, linewidth=2)
```

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income')
```

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True, levels=5, thresh=0.2)
```
