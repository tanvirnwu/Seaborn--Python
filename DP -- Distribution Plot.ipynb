# Distribution Plot
A distribution plot is a graphical representation of the distribution of a dataset. It is a type of histogram that shows the frequency of values in a dataset, but it also includes a smooth curve that represents the estimated probability density function of the data.

In a distribution plot, the x-axis represents the values in the dataset, and the y-axis represents the frequency or density of those values. The smooth curve in the plot is usually generated using a kernel density estimation (KDE) algorithm, which estimates the probability density function of the data.

Distribution plots are commonly used in data analysis and statistics to visualize the distribution of a dataset, and to identify any patterns or outliers. They can be used to compare the distributions of different datasets, or to examine the distribution of a single variable. Examples of distribution plots include the normal distribution, the binomial distribution, and the t-distribution.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

```python
plt.style.use('seaborn')
sns.histplot(data=mart, x='Total')
```

```python
sns.histplot(data=mart, x='Total', binwidth=150)
```

```python
sns.histplot(data=mart, x='Total', bins=20)
```

```python
sns.histplot(data=mart, x='Total', bins=np.arange(0,1100,50))
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', bins=np.arange(0,1100,50))
plt.xticks(np.arange(0,1100,50))
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', bins=np.arange(0,1100,50), kde=True)
plt.xticks(np.arange(0,1100,50))
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', hue='Payment')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', hue='Payment', multiple='stack')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', element='step', fill=False)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', element='poly', fill=True)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', hue='Payment', element='poly', fill=False)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Total', hue='Payment', element='poly', fill=True)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart, x='Payment', shrink=0.7)
```

```python
plt.figure(figsize=(13,5))
sns.histplot(data=mart, x='Total', stat='probability', shrink=0.8)
```
