```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head(5)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', data=mart)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Total', y = 'Product line', hue='Gender', data=mart)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y='Product line', x= 'Total', hue='Gender', data=mart)
```

```python
x = mart['Product line'].sort_values()
x.unique()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'])
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=x)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], capsize=0.2)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], capsize=0.2, ci=None)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], capsize=0.2, ci=None, color='red')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], capsize=0.2, ci=None, palette='BuPu')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], capsize=0.2, ci=None, palette='BuPu', hue_order=['Male', 'Female']
           , saturation=0.4)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y = 'Total', hue='Gender', data=mart
           , order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], capsize=0.2, ci=None, palette='BuPu', hue_order=['Male', 'Female']
           , saturation=0.4, estimator=sum)
```
