# Violin Plot
In Seaborn, a violin plot is a statistical visualization that combines the features of a box plot and a kernel density plot. It shows the distribution of a dataset across different levels of a categorical variable.

A violin plot consists of one or more violins, each representing a category or group of data. Each violin displays a kernel density plot of the data, which shows the probability density of the data at different values along the x-axis. The shape of the violin reflects the density of the data, with wider sections indicating areas where the data are more common.

The central white dot inside each violin represents the median of the data, and the width of the violin at the median is proportional to the interquartile range (IQR) of the data. The thin black lines on either side of the violins represent the range of the data, excluding any outliers.

Violin plots are useful for comparing the distribution of data across multiple categories, and for identifying differences in the shape and spread of the data. They are particularly useful when the data are not normally distributed or when there are multiple modes in the distribution.

Seaborn provides several customization options for violin plots, including changing the color palette, grouping the data by a categorical variable, and changing the orientation of the plot. These options allow you to create more informative and visually appealing visualizations that convey the underlying structure of your data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head(5)
```

```python
sns.violinplot(y='Total', data=mart)
```

```python
sns.violinplot(x='Payment', y='Total', data=mart)
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, palette='BuPu_r')
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, palette='BuPu_r', split=True)
```

```python
plt.figure(figsize=(5,5))
sns.violinplot(y='Payment', x='Total', hue='Gender', data=mart, palette='BuPu_r', split=True)
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, palette='BuPu_r', split=True,
              inner='quartile')
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, palette='BuPu_r', split=True,
              inner='stick')
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, palette='BuPu_r', split=True,
              inner='quartile', bw=.2)
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, palette='BuPu_r', split=True,
              inner='quartile', bw=.2, cut=0)
```
