# Rug Plot
A rug plot is a type of data visualization that is used to represent the distribution of a one-dimensional dataset along a number line. In a rug plot, each data point is represented by a vertical tick mark, or "rug," along the number line, with the density of the tick marks indicating the frequency of the data points at each point along the line.

Rug plots are often used in conjunction with other types of plots, such as histograms or kernel density plots, to provide additional information about the distribution of the data. They can be particularly useful for identifying outliers or for visualizing the distribution of small datasets.

Overall, rug plots provide a quick and easy way to visualize the distribution of a dataset in a compact and informative way.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

```python
plt.figure(figsize=(15,8))
sns.rugplot(data=mart, x='Unit price')
```

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender')
```

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=0.1)
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True)
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=0.1)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=mart, x='Unit price', y='gross income', hue='Gender')
#sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True)
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=0.1)
```

```python
plt.figure(figsize=(15,8))
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True)
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=-0.02, clip_on= False)
```
