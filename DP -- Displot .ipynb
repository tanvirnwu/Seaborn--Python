# Displot 
A displot, short for "distribution plot," is a type of data visualization that displays the distribution of a one-dimensional dataset. It is a combination of a histogram and a kernel density plot (KDE), with the histogram representing the count or density of the data points in each bin, and the KDE providing a smooth estimate of the underlying probability density function.

Displots are often used to visualize the shape, spread, and central tendency of a dataset, as well as to identify any outliers or patterns in the data. They can be created using various programming libraries in Python, such as Seaborn and Matplotlib.

In addition to the histogram and KDE, displots may also include other elements such as rug plots, box plots, or violin plots, to provide additional information about the distribution of the data. They are a useful tool for exploratory data analysis and can help to inform statistical modeling and hypothesis testing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income', 'Branch']]
mart.head(10)
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True)
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True)
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.2})
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.2}, kde_kws={'bw_adjust':9})
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment')
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack')
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack'
           , col='Gender')
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack'
           , col='Gender', row='Branch')
plt.show()
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack'
           , col='Branch', row='Gender')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.displot(data=mart, y='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack'
           , col='Branch', row='Gender')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.displot(data=mart, y='Total', kind='ecdf', hue='Payment'
           , col='Branch', row='Gender')
plt.show()
```
