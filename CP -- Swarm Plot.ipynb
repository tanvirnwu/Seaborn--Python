# Swarm Plot
A swarm plot is a type of plot in Seaborn that displays the distribution of data points as individual points that are arranged in a way that avoids overlap. It is similar to a strip plot, but unlike a strip plot, the points in a swarm plot are adjusted to avoid overlapping each other.

In a swarm plot, each data point is represented as a point along a one-dimensional axis. The points are adjusted along the axis so that they are spaced out from each other, allowing you to see the distribution of the data more clearly.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head(5)
```

```python
sns.swarmplot(y='Total', data=mart)
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='Payment', y='Total', data=mart)
```

```python
plt.figure(figsize=(10,5))

sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart)
```

```python
x=['Cash','Credit card','Ewallet']
plt.figure(figsize=(10,5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart, dodge=True, order=x)
```

```python
x=['Cash','Credit card','Ewallet']
plt.figure(figsize=(15,7))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart, dodge=True, order=x, marker='*', size=7.5)
plt.legend(loc='upper right')
```

```python
x=['Cash','Credit card','Ewallet']
plt.figure(figsize=(15,7))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart, dodge=True, order=x, marker='*', size=7.5)
sns.boxplot(x='Payment', y='Total', data=mart, color='green', boxprops=dict(alpha=0.08))

plt.legend(loc='upper right')
```

```python
x=['Cash','Credit card','Ewallet']
plt.figure(figsize=(15,7))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart, dodge=True, order=x, marker='*', size=7.5, hue_order=['Male','Female'])
sns.violinplot(x='Payment', y='Total', data=mart, palette='spring',  alpha=0.02)

plt.legend(loc='upper right')
```

```python
x=['Cash','Credit card','Ewallet']
plt.figure(figsize=(15,7))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart, dodge=True, order=x, marker='*', size=7.5, hue_order=['Male','Female'])
sns.violinplot(x='Payment', y='Total', data=mart, palette='spring',  alpha=0.02, inner=None)

plt.legend(loc='upper right')
```
