# Box Plot
In Seaborn, a box plot is a statistical visualization that shows the distribution of a dataset. It displays the median, quartiles, and outliers of a dataset in a way that is easy to interpret. The box in the box plot represents the interquartile range (IQR), which is the range between the 25th and 75th percentiles of the data. The line inside the box represents the median of the data, while the whiskers extending from the box show the range of the data, excluding the outliers. Outliers are individual points that fall outside of the whiskers.

Box plots are useful for identifying outliers, comparing the spread and skewness of different datasets, and detecting differences between groups of data. They are often used in statistical analysis and data exploration to gain insights into the distribution of the data.

Seaborn provides several customization options for box plots, including changing the color palette, grouping the data by a categorical variable, and changing the orientation of the plot. These options allow you to create more informative and visually appealing visualizations that convey the underlying structure of your data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head(5)
```

```python
sns.boxplot(x='Total', data=mart, width=0.5)
```

```python
sns.boxplot(y='Total', data=mart, width=0.2)
```

```python
sns.boxplot(x='Payment', y ='Total', data=mart)
```

```python
sns.boxplot(x='Payment', y ='Total', hue='Gender', data=mart)
```

```python
sns.boxplot(x='Payment', y ='Total', hue='Gender', data=mart, showmeans=True)
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Payment', y ='Total', hue='Gender', data=mart, showmeans=True,
            meanprops={"marker":"o", "markerfacecolor":"white",
                       "markersize":"10", "markeredgecolor":"black"})
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Payment', y ='Total', hue='Gender', data=mart, showmeans=True,
            meanprops={"marker":"o", "markerfacecolor":"white",
                       "markersize":"10", "markeredgecolor":"black"},
           palette='BuPu_r', linewidth=4)
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(y='Payment', x='Total', hue='Gender', data=mart, showmeans=True,
            meanprops={"marker":"o", "markerfacecolor":"white",
                       "markersize":"10", "markeredgecolor":"black"},
           palette='BuPu_r', linewidth=4)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=mart)
```
