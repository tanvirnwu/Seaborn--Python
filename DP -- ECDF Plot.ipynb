# ECDF Plot
ECDF stands for Empirical Cumulative Distribution Function. An ECDF plot is a type of graph that is used to visualize the distribution of a one-dimensional dataset. It is constructed by plotting the cumulative proportion of the data points that are less than or equal to each observation on the horizontal axis, and the corresponding observed values on the vertical axis.

In other words, the ECDF plot shows the proportion of data points that are less than or equal to a given value. It can be used to identify important features of the distribution, such as the location of the median or the tails of the distribution.

ECDF plots are useful for comparing the distributions of two or more datasets, as well as for assessing whether a given dataset follows a particular distribution, such as the normal distribution. They are also often used in survival analysis to estimate the survival function, which gives the probability that a subject will survive beyond a certain time.

Overall, the ECDF plot provides a simple and informative way to visualize the distribution of a dataset and is a powerful tool for data exploration and analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

```python
sns.ecdfplot(data=mart,x='gross income')
```

```python
sns.ecdfplot(data=mart,x='gross income', hue='Gender')
```

```python
sns.ecdfplot(data=mart,x='gross income', hue='Gender', stat='count')
```
