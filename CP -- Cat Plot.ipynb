# Cat Plot
In Seaborn, a catplot is a categorical plot that allows you to combine multiple types of plots, such as a scatter plot, a box plot, a bar plot, or a violin plot, into a single figure. It provides a convenient way to visualize the relationship between a numerical variable and one or more categorical variables.

The catplot() function in Seaborn is used to create cat plots. It takes a number of parameters to specify the type of plot to create, the variables to use, and the data to plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head(5)
```

```python
sns.catplot(data=mart, x='Payment', y = 'Total')
```

```python
plt.figure(figsize=(10,5))
sns.catplot(data=mart, x='Payment', y = 'Total', kind='violin')
plt.show()
```

```python
sns.catplot(data=mart, x='Payment', y = 'Total', col='Gender', kind='violin', row='Customer type')
plt.show()
```

```python
sns.catplot(data=mart, x='Payment', y = 'Total', col='Gender', kind='box', row='Customer type', palette='Blues_r')
plt.show()
```

```python
sns.catplot(data=mart, x='Payment', y = 'Total', col='Gender', kind='violin', row='Customer type', palette='Blues_r')
plt.show()
```
