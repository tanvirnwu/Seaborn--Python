# Pair Plot
A pair plot, also known as a scatterplot matrix, is a visualization technique that allows you to explore the relationships between multiple variables at once. It is a grid of scatterplots, where each plot shows the relationship between two variables in the dataset.

In a pair plot, the diagonal plots show the distribution of each variable, while the off-diagonal plots show the scatterplot between each pair of variables. This allows you to quickly identify patterns and relationships between variables in the dataset.

Pair plots are commonly used in exploratory data analysis and can help you identify correlations between variables, outliers, and other patterns in the data. They are often created using data visualization libraries such as Seaborn in Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris, diag_kind='hist')
plt.show()
```

```python
sns.pairplot(data=iris, kind='reg')
plt.show()
```

```python
sns.pairplot(data=iris, diag_kind='kde', hue='species')
plt.show()
```

```python
sns.pairplot(data=iris, x_vars=['sepal_width','petal_width','petal_length'], y_vars='sepal_length')
plt.show()
```

```python
sns.pairplot(data=iris, x_vars=['sepal_width','petal_width','petal_length'], y_vars=['sepal_length','sepal_width'])
plt.show()
```

```python
sns.pairplot(data=iris, corner=True)
plt.show()
```

```python
sns.pairplot(data=iris, corner=True, diag_kws=dict(color='gray',kde=True))
plt.show()
```

```python
sns.pairplot(data=iris, corner=True, diag_kws=dict(color='gray',kde=True),
             plot_kws=dict(color='Green'))
plt.show()
```

```python
sns.pairplot(data=iris, corner=True, diag_kws=dict(color='gray',kde=True),
             plot_kws=dict(color='Green', marker='*', s=50))
plt.show()
```

```python
my_obj1 = sns.pairplot(data=iris, diag_kws=dict(color='gray',kde=True),
             plot_kws=dict(color='Green', marker='*', s=50))

my_obj1.map_upper(sns.kdeplot)
plt.show()
```

```python
my_obj1 = sns.pairplot(data=iris, diag_kws=dict(color='gray',kde=True),
             plot_kws=dict(color='Green', marker='*', s=50))

my_obj1.map_lower(sns.kdeplot)
plt.show()
```
