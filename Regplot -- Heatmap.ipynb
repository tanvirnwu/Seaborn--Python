# Heatmap
A heatmap is a graphical representation of data that uses colors to visualize the relative intensity or density of values within a dataset. Heatmaps are commonly used in data analysis and visualization to represent complex information and patterns in a way that is easy to understand.

Heatmaps typically use a color gradient to represent the intensity of values in the data. For example, a heatmap might use shades of red to represent high values, while shades of blue represent low values. The colors in the heatmap are applied to a grid of cells, where each cell represents a specific value or data point.

Heatmaps can be used for a variety of purposes, such as visualizing the distribution of data in a particular region or identifying patterns in large datasets. They are frequently used in fields such as biology, finance, and marketing to analyze complex data sets and identify trends and patterns that might not be immediately apparent from raw data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('mart_linePlot.xlsx')
mart.columns= mart.columns.str.lower()
mart.head()
```

```python
df = mart.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
df.head()
```

```python
sns.heatmap(data=df)
```

```python
sns.set_style('white')
plt.figure(figsize=(5,5))
sns.heatmap(data=df)
```

```python
sns.set_style('white')
plt.figure(figsize=(5,5))

sns.heatmap(data=df, annot=True)
```

```python
sns.set_style('white')
plt.figure(figsize=(5,7))

sns.heatmap(data=df, annot=True, fmt='.0f')
```

```python
sns.set_style('white')
plt.figure(figsize=(5,6))

sns.heatmap(data=df, annot=True, fmt='.0f',
            annot_kws=dict(size=12, weight= 'bold'))
```

```python
sns.set_style('white')
plt.figure(figsize=(5,6))

sns.heatmap(data=df, annot=True, fmt='.0f',
            annot_kws=dict(size=12, weight= 'bold'),
            linewidth = 0.5, linecolor='white')
```

```python
sns.set_style('white')
plt.figure(figsize=(5,6))

sns.heatmap(data=df, annot=True, fmt='.0f',
            annot_kws=dict(size=12, weight= 'bold'),
            linewidth = 0.5, linecolor='white',
            cmap = 'OrRd_r')
```

```python
mart.corr()
```

```python
sns.heatmap(mart.corr())
```

```python
sns.heatmap(mart.corr(),
            vmin = -1,
            vmax= 1,
            center = 0)
```

```python
sns.heatmap(mart.corr(),
            vmin = -1,
            vmax= 1,
            center = 0,
            cmap = 'OrRd_r',
            annot = True,
            fmt= '.1f',
            annot_kws=dict(size=12, weight='bold'),
            linewidth=2,
            linecolor ='white')
```
