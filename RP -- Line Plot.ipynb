# Line Plot
A line plot, also known as a dot plot or a frequency plot, is a simple graphical representation of a set of data. It is commonly used in statistics to display the distribution of a dataset or to show changes in data over time.

A line plot consists of a horizontal axis (x-axis) and a vertical axis (y-axis). The x-axis represents the range of values of the data, while the y-axis represents the frequency or number of occurrences of each value.

To create a line plot, a dot or mark is placed above the value on the x-axis for each occurrence of that value in the dataset. Then, the dots are connected by a line to show the distribution of the data.

Line plots are useful for identifying patterns and trends in data, as well as for determining the range and distribution of the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart = pd.read_excel('mart_train_sample.xlsx')
mart.head()
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales')
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None)
#Here it shows the mean by default
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, estimator=sum)
#Now as we change the estimator it shows the sum with respect to years
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, estimator=None)
#Now as we change the estimator to None it shows the each data points
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci='sd')
#Now as we change the ci to sd it shows the stand deviation
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=20)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', n_boot=2)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size')
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', estimator=sum,
             palette='rocket_r')
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', estimator=sum,
             palette='rocket_r', style='Outlet_Size', markers=True)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', estimator=sum,
             palette='rocket_r', style='Outlet_Size', markers=True,
             size='Outlet_Size',
             sizes=(1,5))
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', estimator=sum,
              style='Tier', markers=True)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', estimator=sum,
              style='Outlet_Size', markers=True)
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', estimator=sum,
              style='Outlet_Size', markers=True, size='Outlet_Size')
```
