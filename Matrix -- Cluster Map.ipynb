# Cluster Map
A cluster map is a visualization tool that is used to represent clusters or groups of data points based on their similarities or dissimilarities. It is a type of heat map that displays clusters of data points in a two-dimensional space.

Cluster maps are often used in data analysis and machine learning to explore patterns and relationships in large data sets. They can help identify clusters of data points that share common features or characteristics, allowing researchers to better understand the underlying structure of the data.

Cluster maps are typically created using specialized software that applies clustering algorithms to the data set and then generates a heat map based on the resulting clusters. The heat map is then color-coded to represent the degree of similarity or dissimilarity between clusters, with similar clusters assigned similar colors.

Overall, cluster maps are a powerful tool for visualizing complex data sets and uncovering hidden patterns and relationships within the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('mart_linePlot.xlsx')
mart.columns= mart.columns.str.lower()
mart.head()
```

```python
df = mart.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
df.head()
```

```python
sns.clustermap(data=df)
```

```python
sns.clustermap(data=df, col_cluster = False)
```

```python
sns.clustermap(data=df, row_cluster = False, annot =True, fmt='.0f' )
```

```python
sns.clustermap(data=df, row_cluster = False, annot =True, fmt='.0f',
               annot_kws=dict(size=12, weight= 'bold'),
            linewidth = 0.5, linecolor='white')
```

```python
sns.clustermap(data=df, row_cluster = False, annot =True, fmt='.0f',
               annot_kws=dict(size=12, weight= 'bold'),
            linewidth = 0.5, linecolor='white', cmap = 'OrRd_r')
```

```python
sns.clustermap(data=df, row_cluster = False, annot =True, fmt='.0f',
               annot_kws=dict(size=12, weight= 'bold'),
            linewidth = 0.5, linecolor='white', cmap = 'OrRd_r',
              z_score = 1)
```

```python
sns.clustermap(data=df, row_cluster = False, annot =True, fmt='.0f',
               annot_kws=dict(size=12, weight= 'bold'),
            linewidth = 0.5, linecolor='white', cmap = 'OrRd_r',
             # z_score = 1,
              standard_scale = 1)
```
