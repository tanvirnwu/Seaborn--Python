# Joint Plot
A joint plot is a type of data visualization that allows us to compare the relationship between two variables, usually with a scatter plot and marginal histograms or density plots.

In a joint plot, the scatter plot is displayed in the center, showing the relationship between the two variables of interest. The marginal histograms or density plots are shown on the top and right (or left) sides of the scatter plot, respectively, displaying the distribution of each variable separately.

Joint plots can be created using various programming libraries in Python, such as Seaborn and Matplotlib. They are useful for exploring the relationship between two variables, such as assessing the strength and direction of correlation, identifying patterns, and detecting outliers.

In addition to the scatter plot and marginal histograms or density plots, joint plots may also include other elements such as regression lines, confidence intervals, or hexagonal binning, to provide additional information about the relationship between the two variables.

Overall, joint plots are a powerful tool for exploratory data analysis and can help to inform statistical modeling and hypothesis testing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', kind='scatter')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', kind='kde', fill=True)
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', kind='hist')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', kind='hex')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', kind='reg')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', kind='resid')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',  hue='species')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', color='red')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', color='red', hue='species', palette='BuPu')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              joint_kws={'marker':'*', 'color':'black'})
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              joint_kws={'marker':'*', 'color':'black'},
              marginal_kws={'color':'green'})
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              joint_kws={'marker':'*', 'color':'black'},
              marginal_kws={'color':'green', kde=True},
              )
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              joint_kws={'marker':'*', 'color':'black'},
              marginal_kws=dict(color='green', kde=True))
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', 
              joint_kws={'marker':'*', 'color':'black'},
              marginal_kws=dict(color='green', kde=True),
              height=6, ratio=3, space=1)
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length', 
              joint_kws={'marker':'*', 'color':'black'},
              marginal_kws=dict(color='green', kde=True),
              height=6, ratio=3, space=0.5,
              marginal_ticks=True)
```

```python
my_obj1= sns.jointplot(data=iris,x='sepal_length',y='petal_length', 
              joint_kws={'marker':'*', 'color':'red'},
              marginal_kws=dict(color='green', kde=True),
              height=6, ratio=3, space=0.5,
              marginal_ticks=True)

my_obj1.plot_joint(sns.kdeplot)
```

```python
my_obj1= sns.jointplot(data=iris,x='sepal_length',y='petal_length', 
              joint_kws=dict(marker='*', color='red'),
              marginal_kws=dict(color='green', kde=True),
              height=6, ratio=3, space=0.5,
              marginal_ticks=True)

my_obj1.plot_joint(sns.kdeplot)
my_obj1.plot_joint(sns.rugplot, height=0.1, color='red')
```
